```python
import pycaret.datasets as pcdb
data = pcdb.get_data('diabetes')
```

```python
import pycaret.classification as pcc
s = pcc.setup(data, target = 'Class variable',
              session_id = 42)
```

```python
best = pcc.compare_models( )
```

# Evaluating the model

```python
pcc.plot_model(best, plot='confusion_matrix')
```

```python
opt = pcc.optimize_threshold(best, optimize='Accuracy')
pcc.plot_model(opt, plot = 'confusion_matrix')
```

```python
optPrecision = pcc.optimize_threshold(best,
                                      optimize='Precision')
pcc.plot_model(optPrecision, plot = 'confusion_matrix')
```

```python
optF1 = pcc.optimize_threshold(best, optimize='F1')
pcc.plot_model(optF1, plot='confusion_matrix')
```

```python
optKappa = pcc.optimize_threshold(best, optimize='Kappa')
pcc.plot_model(optKappa, plot='confusion_matrix')
```

```python
best3 = pcc.compare_models(n_select=3, sort='F1')
stack = pcc.stack_models(best3, optimize='F1')
opstackF1 = pcc.optimize_threshold(stack, optimize='F1')
pcc.plot_model(opstackF1, 'confusion_matrix')
```
